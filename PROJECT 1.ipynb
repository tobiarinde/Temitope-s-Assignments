```python
# Chart to show Product performance
sns.histplot(data=cardio, x='Product')
plt.xlabel('Product Offered')
plt.ylabel('Range')
plt.title('Histogram Showing Gym Product Offered')
plt.show()
```

```python
# The chart above shows the Products offered to customers and their market performance
```

```python
# Chart to show Customer Tips
sns.countplot(data=cardio, x='Age')
plt.xlabel('Client Age Group')
plt.ylabel('Range')
plt.title('Histogram Showing Customer Age Range')
plt.show()
```

```python
# The chart below shows the Age groups that patronize the Gym and also ranks the Age Group with highest patronage/usage
```

```python
# Chart to show Gender Rank
sns.countplot(data=cardio, x ='Gender')
plt.xlabel('Gender')
plt.ylabel('Range')
plt.title('Boxplot Showing Gender')
plt.show()
```

```python
# The chart above shows that more male visit the gym than female
```

```python
# Chart to show Marital Status ratio
sns.countplot(data=cardio, x='MaritalStatus')
plt.xlabel('Marital Status')
plt.ylabel('Range')
plt.title('Histogram showing Marital Status')
plt.show()
```

```python
#The chart above shows that people who visit the gym come in partnered more often than single (individual)
```

```python
cardio.columns
corr_mat=cardio[cardio.columns].corr()
corr_mat
```

```python
# This plot is to show whether any parameter will affect the function model of the Dataset but it shows none is correlated
plt.figure()
sns.heatmap(data = corr_mat, annot = True, vmin = -1, vmax = 1, cmap = 'Spectral');
```

```python
# Chart to show Product performance within Age groups
sns.catplot(data=cardio, x='Product', y='Age', kind='box')
plt.xticks(rotation=90)
plt.ylabel('Age')
plt.xlabel('Product')
plt.title('Cardio CategoricalPlot')
plt.show()
```

```python
# The chart above shows the product that performs greatest amongst the Age ranges
```

```python
# Chart to show usage each gender pulls in their ages
sns.scatterplot(data=cardio, x='Usage', y='Age', hue='Gender')
plt.xlabel('Usage')
plt.ylabel('Age Range')
plt.title('Gym Usage ScatterPlot')
plt.show()
```

```python
# The chart above shows the Age group and how often they use the gym weekly 
```

```python
# Chart to show the income earned from their Marital Status within an Age bracket
sns.scatterplot(data=cardio, x='Age', y='Income', hue='MaritalStatus')
plt.xlabel('Age')
plt.ylabel('Income Earned')
plt.title('Earnings ScatterPlot')
plt.show()
```

```python
# The chart above shows how much income is generated between the Single and the Partnered Gym Users
```

```python
# Chart to show Income earned between Genders in their Age Group
sns.scatterplot(data=cardio, x='Age', y='Income', hue='Gender')
plt.xlabel('Age')
plt.ylabel('Income Earned')
plt.title('Earnings ScatterPlot')
plt.show()
```

```python
# The Chart above shows how much income is generated between the Male and the Female
```

```python
# Chart to show fitness between genders with Age groups
sns.scatterplot(data=cardio, x='Age', y='Fitness', hue='Gender')
plt.xlabel('Age')
plt.ylabel('Fitness')
plt.title('Gym Fitness ScatterPlot')
plt.show()
```

```python
# This chart shows that as the men stay more fit than women even as they age
```

```python
# Chart to show fitness as regards to usage within genders
sns.scatterplot(data=cardio, x='Usage', y='Fitness', hue='Gender')
plt.xlabel('Usage')
plt.ylabel('Fitness')
plt.title('Gym Usage/Fitness ScatterPlot')
plt.show()
```

```python
# This chart above shows that the high usage of the male users resulted to their high fitness compared to the female
```
